```python
from bs4 import BeautifulSoup
import requests
import re
content = requests.get('https://en.wikipedia.org/wiki/Gallery_of_sovereign_state_flags')

soup = BeautifulSoup(content.text, 'lxml')
images = [x.get('src')[2:] for x in soup.find_all('img', {'class': 'mw-file-element'}) if 'Flag' in x.get('src') ] 
flags_folder = 'flags'
for x in images:
    name = re.findall(r'Flag_of_(.*)\.svg', x)[0].split('.')[0].split('_%')[0].replace('_', ' ').lower()
    with open(f'{flags_folder}/{name}.png', 'wb') as f:
        f.write(requests.get(f'https://{x}').content)
        
        
    
```

```python
#change the name of every flag file into the same name but with the first letter being capital
import os
for filename in os.listdir('flags'):
    os.rename(f'flags/{filename}', f'flags/{filename.title()}')
```
